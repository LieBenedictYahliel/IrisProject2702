# Iris Project

Iris spesies Classification

```python
# Import libaries
import pandas as pd
import numpy as np

# Sklearn
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler

# Joblib
import joblib
```

```python
# import data
data=load_iris(as_frame=True)
x=data.data
y=data.target

data.target_names
```

```python
display(x,y)
```

## Classifier

```python
clf=Pipeline([
    ("scaler", MinMaxScaler()),
    ("classifier", LogisticRegression())
])

clf.fit(x,y)
```

```python
clf.score(x,y)
```

```python
# Save model
joblib.dump(clf,"model.joblib")
```

Keterbatasan: 
- Keterbatasan UI kompleks: Streamlit cocok untuk dashboard sederhana sampai menengah, tapi sulit atau tidak fleksibel untuk UI yang sangat kompleks dan interaktif seperti aplikasi web full-featured.
- Kustomisasi terbatas: Pilihan styling dan layout default Streamlit sangat terbatas dibanding framework web tradisional (React, Vue), sehingga sulit bikin tampilan yang sangat spesifik atau branding kuat.
- Performa untuk skala besar: Aplikasi Streamlit cenderung berat dan lambat jika data atau pengguna sangat banyak, karena setiap interaksi biasanya reload sebagian besar script.
- State management minimal: Pengelolaan state aplikasi kurang matang, sehingga sulit buat aplikasi dengan banyak interaksi atau sesi pengguna yang kompleks.
- Tidak mendukung routing/page multi-view dengan baik: Untuk aplikasi multi halaman, Streamlit masih terbatas, meski ada workaround tapi bukan native dan kurang elegan.
- Kurangnya kontrol penuh backend: Karena fokus ke front-end cepat, developer punya kontrol terbatas terhadap backend, middleware, atau API kompleks.
- Bergantung pada Python saja: Jika ingin kombinasi bahasa atau ekosistem lain, tidak fleksibel karena semuanya harus di Python.
